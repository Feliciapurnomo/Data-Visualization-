```python
import pandas as pd
import seaborn as sns
```

```python
# Load dataset
df_age = pd.read_csv('/content/gdrive/MyDrive/Dibimbing/2/Enter_korea_by_age.csv')
df_age
```

```python
df_age.info()
```

```python
# draw barchart of visitor
sns.catplot(data=df_age,
            x='visitor',
            y='nation',
            kind='bar',
            ci=None,
            height=10, 
            aspect=2.5)
```

```python
data_merged = df_age.drop(columns=["date","visitor","growth","share"])
data_merged
```

```python
data_tidy = pd.melt(data_merged,
                    id_vars='nation',
                    var_name='age',
                    value_name='total_visitor')

data_tidy
```
